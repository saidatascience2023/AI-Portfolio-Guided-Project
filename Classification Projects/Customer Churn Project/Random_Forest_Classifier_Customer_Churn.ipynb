```python
import math

def plot_categorical_features(df, max_unique=30):
    # Select categorical columns with less than 15 unique values
    categorical_columns = [
        col for col in df.select_dtypes(include=['object', 'category']).columns
        if df[col].nunique() < max_unique
    ]

    # Get the number of selected categorical columns
    num_columns = len(categorical_columns)

    # If no suitable columns are found, print a warning and exit
    if num_columns == 0:
        print(f"No categorical columns with less than {max_unique} unique values.")
        return

    # Determine the number of rows and columns for the plot layout
    num_rows = math.ceil(num_columns / 3)
    num_cols = min(3, num_columns)

    # Create a grid for subplots
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, num_rows * 5))
    axes = axes.flatten()

    for i, col in enumerate(categorical_columns):
        # Plot bar chart for each categorical column
        df[col].value_counts().plot.bar(
            ax=axes[i],
            color='lightblue',
            edgecolor='black'
        )
        axes[i].set_title(f'Distribution of {col}')
        axes[i].set_ylabel('Count')
        axes[i].set_xlabel('Categories')

    # Remove any extra subplots if the number of columns is not a perfect multiple of 3
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
```

# Handle the Outliers

```python
import numpy as np
from sklearn.preprocessing import RobustScaler

# Function to detect and cap outliers based on IQR (only for training data)
def cap_outliers_iqr(train, test, column):
    Q1 = train[column].quantile(0.25)
    Q3 = train[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Apply capping using training data bounds
    train[column] = np.where(train[column] > upper_bound, upper_bound, train[column])
    train[column] = np.where(train[column] < lower_bound, lower_bound, train[column])

    test[column] = np.where(test[column] > upper_bound, upper_bound, test[column])
    test[column] = np.where(test[column] < lower_bound, lower_bound, test[column])

# Apply outlier handling to both 'SeniorCitizen' and 'Missing_TotalCharges'
cap_outliers_iqr(X_train, X_test, "SeniorCitizen")
cap_outliers_iqr(X_train, X_test, "Missing_TotalCharges")

# Apply RobustScaler (after outlier handling)
scaler = RobustScaler()

# Scaling columns after outlier handling
X_train["SeniorCitizen"] = scaler.fit_transform(X_train[["SeniorCitizen"]])
X_test["SeniorCitizen"] = scaler.transform(X_test[["SeniorCitizen"]])

X_train["Missing_TotalCharges"] = scaler.fit_transform(X_train[["Missing_TotalCharges"]])
X_test["Missing_TotalCharges"] = scaler.transform(X_test[["Missing_TotalCharges"]])
```

# Logistic Regression Model Implementation Using Pipelines and ColumnTansformers

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression


def tenure_based_imputation(df):
    # Ensure tenure is numeric
    df['tenure'] = pd.to_numeric(df['tenure'], errors='coerce')

    # Compute tenure-based mean
    total_charges_mean_by_tenure = df.groupby('tenure')['TotalCharges'].transform('mean')

    # Fill missing values with the tenure-based mean
    df['TotalCharges'] = df['TotalCharges'].fillna(total_charges_mean_by_tenure)

    # Fallback: Fill remaining NaNs with overall median
    df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)

    return df  # Return as DataFrame to maintain columns and indices

# Wrap Function in FunctionTransformer
group_imputer = FunctionTransformer(tenure_based_imputation, validate=False)


# **Step 2: Numeric Data Preprocessing Pipeline**
num_pipeline = Pipeline([
    ('group_imputer', group_imputer),  # Custom tenure-based imputation
    ('scaler', StandardScaler())  # Standardize numeric features
])

# **Step 3: Categorical Data Preprocessing Pipeline**
cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Fill missing categorical values
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))  # One-hot encode categorical features
])

# **Step 4: ColumnTransformer to Apply Preprocessing Pipelines**
col_transformer = ColumnTransformer([
    ('num', num_pipeline, num_cols),  # Apply num_pipeline to numeric columns
    ('cat', cat_pipeline, cat_cols)   # Apply cat_pipeline to categorical columns
], verbose_feature_names_out=False)

# **Step 5: Define the Final Model Pipeline with Logistic Regression**
model_pipeline_lgr = Pipeline([
    ('preprocessor', col_transformer),
    ('classifier', LogisticRegression(max_iter=1000,random_state=42))  # Logistic Regression Model
])

```

## What is Precision

1. Precision (Positive Predictive Value)
Formula:

Precision-Recall-F1-Score-04-03-2025_11_05_AM.png

**TP (True Positives):** Customers correctly predicted as "Churn".

**FP (False Positives):** Customers incorrectly predicted as "Churn" when they actually stayed.

Interpretation:

**High Precision** → When the model says a customer will churn, it is very likely correct.

**Low Precision** → Many customers predicted as "Churn" are actually "No Churn" (false alarms).

**When is Precision Important?**

If the cost of wrongly predicting a customer will churn (False Positive) is high (e.g., offering expensive retention discounts unnecessarily), we want high precision.

## What is Recall

2. Recall (Sensitivity or True Positive Rate)
Formula:

Precision-Recall-F1-Score-04-03-2025_11_08_AM.png

**FN (False Negatives):** Customers predicted as "No Churn" but actually churned.

Interpretation:

**High Recall** → The model captures most of the actual churned customers.

**Low Recall** → Many churned customers are missed (False Negatives).

**When is Recall Important?**

If missing a churning customer is costly (e.g., a high-value customer leaving without intervention), we prioritize high recall.

## F1 Score

3. F1-Score (Harmonic Mean of Precision & Recall)
Formula:

Precision-Recall-F1-Score-04-03-2025_11_10_AM.png

F1-score balances precision and recall.

High F1-score means a model is good at both avoiding false positives and false negatives.

**When to use F1-Score?**

When both False Positives and False Negatives are costly, we use F1-score for a balanced metric.

In most churn prediction scenarios, Recall is prioritized because missing churners is more damaging than falsely predicting churn.

F1-score is a good overall metric if both false positives and false negatives matter.
